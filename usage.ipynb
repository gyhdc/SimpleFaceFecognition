```python
import sys
sys.path.append(r"D:\Desktop\深度学习\11.大三下课设\人脸识别")
from face_recognizition import Recognizer
from opencv_module import OpencvModule
```

```python
recognizer=Recognizer()
OpencvModule=OpencvModule()
```

```python
#opencv打开摄像头，并人脸识别登录
#opencv打开摄像头，并人脸识别登录
import cv2
cap = cv2.VideoCapture(0)
cnt=0
tar=20

while(True):
    # Capture frame-by-frame
    ret, frame = cap.read()
    if cnt<=3:#缓冲3帧
        cv2.imshow('frame',OpencvModule.target_detect(frame))
        cnt+=1
        continue
    if True:
        res=recognizer.recognition_target(frame,threshold=0.8)#识别
        frame=res['frame']
        user=res['user']
        conf=res['conf']
    if user is not None:
        cnt+=1
    cv2.imshow('frame',frame)
    if cnt==tar:
        print(f"用户 {user} 登陆成功 置信度 {conf}")
        break
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
